# 2. Which sport is having most number of Gold Medals so far? (Top 5)


```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")      #read the olympic dataset
data1=[]                            #initialsing an empty list to store the data
df1 = (df[df['Medal'] == 'Gold'])   #storing the gold medals coloumn inside a variable 
for Sport in df1['Sport'].unique(): #will give the unique number of Sports
    data1.append([Sport, len(df1[df1['Sport'] == Sport])])   #appending the data inside the list and further plotting 
pd.DataFrame(data1,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

# 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")                #read the olympic dataset 
data7 = []                                    #initialsing an empty list to store the data
for Event in df['Event'].unique():            #will give the unique events
    data7.append([Event, len(df[df['Event'] == Event])])     #appending the data inside the list and further requirements
pd.DataFrame(data7,columns = ['Event','Person']).sort_values(by='Person', ascending=False).head().plot(x = 'Event', y = 'Person', kind = 'bar', figsize = (5,5))
```

# 8. Which sport is having most female Gold Medalists? (Top 5)


```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")                #read the olympic dataset 
data8 = []                                    #initialsing an empty list to store the data
df3 = (df[df['Medal'] == 'Gold'])             #initialising a variable and storing the data of all gold medalist
df4 = df3[df3['Gender'] == 'Women']           #initialising a variable and storing the data of all female gold medalist
for Sport in (df4['Sport'].unique()):         #will give the unique sports events
    data8.append((Sport , len(df4[df4['Sport'] == Sport])))    #appending the data of all female gold medalists and plotting it further
pd.DataFrame(data8,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```
